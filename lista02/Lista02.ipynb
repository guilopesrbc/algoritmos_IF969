```python
class Node:
    def __init__(self, value):
        self.value = value
        self.parent = None
        self.rightChild = None
        self.leftChild = None
        self.fb = 0


class AVLtree:
    def __init__(self):
        self.root = None
        self.in_order_print = []
        self.inbalance = []

    def search(self, txt):
        root = self.root
        while root and root.value != txt:
            if txt < root.value:
                root = root.leftChild
                if root is None:
                    return None
            else:
                root = root.rightChild
                if root is None:
                    return None
        return root

    def min_tree(self, root):
        while root.leftChild:
            root = root.leftChild
        return root

    def max_tree(self, root):
        while root.rightChild:
            root = root.rightChild
        return root

    def height_tree(self, root):
        if root:
            leftHeight = self.height_tree(root.leftChild)
            rightHeight = self.height_tree(root.rightChild)
            # retorna o maximo entre altura da arvore direita e esquerda + 1 pra contar a raiz como se fosse altura 1
            return max(leftHeight, rightHeight) + 1
        else:
            return 0
    def in_order(self, root):
        if root:
            if root.leftChild:
                self.in_order(root.leftChild)
            self.in_order_print.append(root.value)
            if root.rightChild:
                self.in_order(root.rightChild)

    def leftRotation(self, root):
        aux = root.parent
        if aux == self.root:
            self.root = root

        # se tiver filho a esquerda
        if root.leftChild:

            # vendo de qual lado o pai é filho do pai dele
            if root.parent and root.parent.parent:
                if root.parent != self.root and root.parent.parent.rightChild == root.parent:
                    root.parent.parent.rightChild = root
                elif root.parent != self.root and root.parent.parent.leftChild == root.parent:
                    root.parent.parent.leftChild = root

            # set dos filhos
            root.leftChild.parent = root.parent
            root.parent.rightChild = root.leftChild
            root.leftChild = root.parent
            root.parent = root.parent.parent
            aux.parent = root

        else:
            # vendo de qual lado o pai é filho do pai dele
            if root.parent and root.parent.parent:
                if root.parent != self.root and root.parent.parent.rightChild == root.parent:
                    root.parent.parent.rightChild = root
                elif root.parent != self.root and root.parent.parent.leftChild == root.parent:
                    root.parent.parent.leftChild = root

            # set dos filhos
            root.parent.rightChild = None
            root.leftChild = root.parent
            root.parent = root.parent.parent
            aux.parent = root

    def rightRotation(self, root):
        aux = root.parent
        if aux == self.root:
            self.root = root

        # se tiver filho a direita
        if root.rightChild:

            # vendo de qual lado o pai é filho do pai dele
            if root.parent and root.parent.parent:
                if root.parent != self.root and root.parent.parent.rightChild == root.parent:
                    root.parent.parent.rightChild = root
                elif root.parent != self.root and root.parent.parent.leftChild == root.parent:
                    root.parent.parent.leftChild = root

            # set dos filhos
            root.rightChild.parent = root.parent
            root.parent.leftChild = root.rightChild
            root.rightChild = root.parent
            root.parent = root.parent.parent
            aux.parent = root

        else:
            # vendo de qual lado o pai é filho do pai dele
            if root.parent and root.parent.parent:
                if root.parent != self.root and root.parent.parent.rightChild == root.parent:
                    root.parent.parent.rightChild = root
                elif root.parent != self.root and root.parent.parent.leftChild == root.parent:
                    root.parent.parent.leftChild = root

            # set dos filhos
            root.parent.leftChild = None
            root.rightChild = root.parent
            root.parent = root.parent.parent
            aux.parent = root

    def leftleftRotation(self, root):
        self.leftRotation(root)
        self.leftRotation(root)

    def leftRightRotation(self, root):
        self.leftRotation(root)
        self.rightRotation(root)

    def rightrightRotation(self, root):
        self.rightRotation(root)
        self.rightRotation(root)

    def rightleftRotation(self, root):
        self.rightRotation(root)
        self.leftRotation(root)

    def balance(self, root):
        if root:
            root.fb = self.height_tree(root.leftChild) - self.height_tree(root.rightChild)
            if root.fb < -1 or root.fb > 1:
                self.inbalance.append(root)
            if root.leftChild:
                self.balance(root.leftChild)
            if root.rightChild:
                self.balance(root.rightChild)

    def verify_rotations(self, root):
        if root.fb < -1:
            # rotação direita esquerda
            if root.rightChild and (root.rightChild.fb == 1 or root.rightChild.fb == 0):
                self.rightleftRotation(root.rightChild.leftChild)
            # rotação esquerda esquerda
            elif root.rightChild and (root.rightChild.fb == -1 or root.rightChild.fb == 0):
                self.leftleftRotation(root.rightChild.rightChild)
        elif root.fb > 1:
            # rotação esquerda direita
            if root.leftChild and (root.leftChild.fb == -1 or root.leftChild.fb == 0):
                self.leftRightRotation(root.leftChild.rightChild)
            # rotação direita direita
            elif root.leftChild and (root.leftChild.fb == 1 or root.leftChild.fb == 0):
                self.rightrightRotation(root.leftChild.leftChild)
        self.balance(self.root)

    def remove_no_child(self, root):
        # verifica se é a raiz
        if not root.parent:
            self.root = None
        else:
            # verifica se é filho a esquerda ou a direita
            if root.parent.leftChild == root:
                root.parent.leftChild = None
            elif root.parent.rightChild == root:
                root.parent.rightChild = None

    def remove_one_child(self, root):
        # verifica se é a raiz
        if not root.parent:
            # verifica se o filho é a esquerda
            if root.leftChild:
                self.root = root.leftChild
                root.leftChild.parent = None
            # é filho a direita
            else:
                self.root = root.rightChild
                root.rightChild.parent = None
        else:
            # verifica se o filho é a esquerda
            if root.leftChild:
                # verifica se o nó é filho a esquerda
                if root.parent.leftChild == root:
                    root.parent.leftChild = root.leftChild
                    root.leftChild.parent = root.parent
                else:
                    # o nó é filho a direita
                    root.parent.rightChild = root.leftChild
                    root.leftChild.parent = root.parent
            else:
                # é filho a direita
                # verifica se o nó é filho a esquerda
                if root.parent.leftChild == root:
                    root.parent.leftChild = root.rightChild
                    root.rightChild.parent = root.parent
                else:
                    # o nó é filho a direita
                    root.parent.rightChild = root.rightChild
                    root.rightChild.parent = root.parent

    def remove_two_childs(self, root):
        parent_successor = self.min_tree(root.rightChild).parent
        successor = self.min_tree(root.rightChild)
        # verifica se é raiz
        if not root.parent:
            # caso o sucessor seja o filho da raiz
            if parent_successor == root:
                successor.leftChild = root.leftChild
                successor.parent = None
                successor.leftChild.parent = successor
            else:
                # verifica se o sucessor é filho a esquerda ou direita
                if parent_successor.leftChild == successor:
                    parent_successor.leftChild = None
                    successor.parent = None
                    root.leftChild.parent = successor
                    root.rightChild.parent = successor
                else:
                    parent_successor.rightChild = None
                    successor.parent = None
                    root.leftChild.parent = successor
                    root.rightChild.parent = successor

            # define os filhos do sucessor
            if root.rightChild == successor:
                successor.rightChild = None
            else:
                successor.rightChild = root.rightChild
            successor.leftChild = root.leftChild
            self.root = successor

        else:
            # verifica se o nó é filho a esquerda ou direita
            if root.parent.leftChild == root:
                root.parent.leftChild = successor
                successor.parent = root.parent
                root.leftChild.parent = successor
                root.rightChild.parent = successor

            else:
                root.parent.rightChild = successor
                successor.parent = root.parent
                root.leftChild.parent = successor
                root.rightChild.parent = successor

            # verifica se o sucessor é filho a esquerda ou direita
            if parent_successor.leftChild == successor:
                parent_successor.leftChild = None
            else:
                parent_successor.rightChild = None

            # define os filhos esquerda e direita do sucessor
            successor.leftChild = root.leftChild
            successor.rightChild = root.rightChild

    def insert(self, txt):
        if self.search(txt) is not None:
            print(f'{txt} JA EXISTE')
        new_root = Node(txt)
        # caso não há nó inserido
        if not self.root:
            self.root = new_root
        else:
            # inserção
            aux_root = self.root
            while True:
                # valor menor, vai para a esquerda
                if new_root.value < aux_root.value:
                    # set dos filhos
                    if not aux_root.leftChild:
                        aux_root.leftChild = new_root
                        new_root.parent = aux_root
                        break
                    aux_root = aux_root.leftChild
                # valor maior, vai para a direita
                else:
                    # set dos filhos
                    if not aux_root.rightChild:
                        aux_root.rightChild = new_root
                        new_root.parent = aux_root
                        break
                    aux_root = aux_root.rightChild

            # faz o balanceamento
            self.balance(self.root)
            while len(self.inbalance) != 0:
                aux = self.inbalance[0]
                if aux.fb < -1 or aux.fb > 1:
                    self.verify_rotations(aux)
                self.inbalance.pop(0)

        print(f'{txt} INSERIDO')

    def remove(self, txt):
        root = self.search(txt)
        if not root:
            return print(f'{txt} NAO ENCONTRADO')

        else:
            # caso não tenha filhos PRIMEIRO CASO
            if not root.leftChild and not root.rightChild:
                self.remove_no_child(root)

            # caso só possua um filho SEGUNDO CASO
            elif (not root.leftChild and root.rightChild) or (root.leftChild and not root.rightChild):
                self.remove_one_child(root)

            # caso possua dois filhos TERCEIRO CASO
            elif root.leftChild and root.rightChild:
                self.remove_two_childs(root)

            # faz o balanceamento
            self.balance(self.root)
            while len(self.inbalance) != 0:
                aux = self.inbalance[0]
                if aux.fb < -1 or aux.fb > 1:
                    self.verify_rotations(aux)
                self.inbalance.pop(0)

        print(f'{txt} DELETADO')

    # DESAFIO_______________________________________________
    def ver(self):
        print('EM ORDEM:')
        # chama o print dos nós em ordem com a mensagem. PS: coloquei uma função diferente da função in_order para não atrapalhar no output do dikastis
        self.ver_order(self.root)
    def ver_order(self, root):
        if root:
            if root.leftChild:
                self.ver_order(root.leftChild)
            if root.leftChild and root.rightChild:
                print(f'{root.value} ajuda {root.leftChild.value} e {root.rightChild.value}')
                print('_____________________________________')
            elif root.leftChild:
                print(f'{root.value} ajuda {root.leftChild.value}')
                print('_____________________________________')
            elif root.rightChild:
                print(f'{root.value} ajuda {root.rightChild.value}')
                print('_____________________________________')
            else:
                print(f'{root.value} ajuda ninguém')
                print('_____________________________________')
            if root.rightChild:
                self.ver_order(root.rightChild)
    # função ver para um nó especifico
    def ver_node(self, txt):
        root = self.search(txt)
        print('___________________________________')
        if root:
            if root.leftChild and root.rightChild:
                print(f'{root.value} ajuda {root.leftChild.value} e {root.rightChild.value}')
                print('_____________________________________')
            elif root.leftChild:
                print(f'{root.value} ajuda {root.leftChild.value}')
                print('_____________________________________')
            elif root.rightChild:
                print(f'{root.value} ajuda {root.rightChild.value}')
                print('_____________________________________')
            else:
                print(f'{root.value} ajuda ninguém')
                print('_____________________________________')
        else:
            print('esse nó não existe')
    #____________________________________________________________

tree = AVLtree()
end = False
while not end:
    list_of_messages = input().split()
    if list_of_messages[0] == 'INSERIR':
        tree.insert(list_of_messages[1])

    elif list_of_messages[0] == 'DELETAR':
        tree.remove(list_of_messages[1])

    elif list_of_messages[0] == 'MINIMO':
        if tree.root:
            print(f'MENOR: {tree.min_tree(tree.root).value}')
        else:
            print('ARVORE VAZIA')

    elif list_of_messages[0] == 'MAXIMO':
        if tree.root:
            print(f'MAIOR: {tree.max_tree(tree.root).value}')
        else:
            print('ARVORE VAZIA')

    elif list_of_messages[0] == 'ALTURA':
        if tree.root:
            print(f'ALTURA: {tree.height_tree(tree.root)}')
        else:
            print('ALTURA: 0')

    elif list_of_messages[0] == 'FIM':
        if tree.root:
            tree.in_order(tree.root)
            print(' '.join(tree.in_order_print))
            tree.in_order_print = []
        else:
            print('ARVORE VAZIA')
        end = True


```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.pai = None
        self.direita = None
        self.esquerda = None
        self.cl = 0
        self.pl = 0

class Arvore_binaria:
    def __init__(self):
        self.raiz = None

    def add(self, num):
        new_node = Node(num)
        actual_node = self.raiz
        aux_node = None

        # não há nenhum nó ainda foi inserido
        if actual_node == None:
            self.raiz = new_node
            new_node.pl = new_node.cl
            print(new_node.cl)

        else:
            # verifica se o nó vai para a esquerda ou direita
            while actual_node != None:
                aux_node = actual_node

                # o nó a ser inserido vai para a esquerda
                if new_node.value < actual_node.value:
                    actual_node = actual_node.esquerda

                # o nó a ser inserido vai para a direita
                else:
                    actual_node = actual_node.direita

                new_node.cl += 1
            new_node.pl = new_node.cl
            # define ponteiros
            if new_node.value < aux_node.value:
                aux_node.esquerda = new_node
                new_node.pai = aux_node
                print(new_node.cl)

            else:
                aux_node.direita = new_node
                new_node.pai = aux_node
                print(new_node.cl)

    def rotacaoEsquerda(self, node):
        aux = node.pai
        # se tiver filho a esquerda
        if node.esquerda is not None:
            # vendo de qual lado o pai é filho do pai dele
            if node.pai != self.raiz and node.pai.pai.direita == node.pai:
                node.pai.pai.direita = node
            elif node.pai != self.raiz and node.pai.pai.esquerda == node.pai:
                node.pai.pai.esquerda = node

            node.esquerda.pai = node.pai
            node.pai.direita = node.esquerda
            node.esquerda = node.pai
            node.pai = node.pai.pai
            aux.pai = node

        else:
            if node.pai != self.raiz and node.pai.pai.direita == node.pai:
                node.pai.pai.direita = node
            elif node.pai != self.raiz and node.pai.pai.esquerda == node.pai:
                node.pai.pai.esquerda = node

            node.pai.direita = None
            node.esquerda = node.pai
            node.pai = node.pai.pai
            aux.pai = node

    def rotacaoDireita(self, node):
        aux = node.pai
        # se tiver filho a direita
        if node.direita is not None:
            # vendo de qual lado o pai é filho do pai dele
            if node.pai != self.raiz and node.pai.pai.direita == node.pai:
                node.pai.pai.direita = node
            elif node.pai != self.raiz and node.pai.pai.esquerda == node.pai:
                node.pai.pai.esquerda = node

            node.direita.pai = node.pai
            node.pai.esquerda = node.direita
            node.direita = node.pai
            node.pai = node.pai.pai
            aux.pai = node

        else:
            if node.pai != self.raiz and node.pai.pai.direita == node.pai:
                node.pai.pai.direita = node
            elif node.pai != self.raiz and node.pai.pai.esquerda == node.pai:
                node.pai.pai.esquerda = node

            node.pai.esquerda = None
            node.direita = node.pai
            node.pai = node.pai.pai
            aux.pai = node

    def search(self, num):
        actual_node = self.raiz
        while actual_node.value != num:
            if num < actual_node.value:
                actual_node = actual_node.esquerda
                if actual_node is None:
                    print(-1)
                    return None
            else:
                actual_node = actual_node.direita
                if actual_node is None:
                    print(-1)
                    return None
        return actual_node

    def rotacoes(self,actual_node):
        while self.raiz != actual_node:
            # casos de rotação
            # rotação a esquerda
            if actual_node.pai.direita == actual_node:
                if actual_node.pai == self.raiz:
                    self.rotacaoEsquerda(actual_node)
                    self.raiz = actual_node
                else:
                    self.rotacaoEsquerda(actual_node)
            # rotação a direita
            else:
                if actual_node.pai == self.raiz:
                    self.rotacaoDireita(actual_node)
                    self.raiz = actual_node
                else:
                    self.rotacaoDireita(actual_node)

    def sch(self, num):
        # busca o nó
        actual_node = self.search(num)

        # realiza as rotações
        if actual_node is not None:
            self.rotacoes(actual_node)
        else:
            return None
        # printa o pl e atualiza o cl de cada nó
        self.raiz = actual_node
        print(actual_node.pl)
        actual_node.cl = 0
        self.atualiza_cl(actual_node)


    def atualiza_cl(self, node):
        if node is not None:
            if node.value is not self.raiz.value:
                node.cl = node.pai.cl + 1
            node.pl = node.cl
            self.atualiza_cl(node.direita)
            self.atualiza_cl(node.esquerda)

arvore = Arvore_binaria()
try:
    while True:
        entrada = input().split()
        if entrada[0] == "ADD":
            arvore.add(int(entrada[1]))
        elif entrada[0] == "SCH":
            arvore.sch(int(entrada[1]))
except:
    pass
```
