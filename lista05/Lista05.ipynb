```python
class Graph:
    # construção do grafo usando lista de adjacência
    def __init__(self, number_of_vertexes, boost):
        self.number_of_vertexes = number_of_vertexes
        self.graph = []
        # construção da lista de adjacência
        for i in range(number_of_vertexes):
            self.graph.append([])
        self.boost = boost

    # constroi as ligações direcionadas
    def connection(self, vertex1, vertex2):
        self.graph[vertex1].append(vertex2)

    # busca em largura
    def bfs(self, vertex):
        # lista dos vertices marcados
        self.marked = [vertex]
        # lista dos nós alcançados
        self.range = []
        # faz a marcação e visitação em largura
        self.marking(vertex)
        # função para definir qual publico será alcançado de acordo com o boost pago
        self.boost_times(vertex)

    # realiza a marcação e visita em largura
    def marking(self, vertex):
        # exclui o marcado e visita os seus adjacentes
        self.marked.pop(0)
        # adiciona na lista de alcançados
        self.range.append(vertex)
        # adiciona à lista dos marcados os vertices adjacentes
        for value in self.graph[int(vertex)]:
            if value not in self.marked and value not in self.range:
                self.marked.append(value)
        # realiza a visita com o proximo vertice na lista de marcação
        for i in self.marked:
            self.marking(i)

    # função para definição da lista de seguidores atingidos
    def boost_times(self, vertex):
        self.range.pop(0)
        # define quantos seguidores serão atingidos além dos seguidores free
        count = 0
        while self.boost > 5.25:
            self.boost -= 5.25
            count += 1
        # exclui os seguidores que não serão atingidos
        count += len(self.graph[int(vertex)])
        for i in range(count, len(self.range)):
            self.range.pop(count)
        print(self.range)

number_of_vertexes = int(input())
id = input()
boost = float(input())

# programa da questão
g = Graph(number_of_vertexes, boost)
for i in range(number_of_vertexes):
    list_of_inputs = input().split()
    vertex = int(list_of_inputs[0])
    for j in range(2, len(list_of_inputs)):
        g.connection(vertex, list_of_inputs[j])
# realiza a busca em largura
g.bfs(id)
```

```python
class Graph:
    # construção do grafo usando matriz de adjacência
    def __init__(self, number_of_vertexes):
        self.number_of_vertexes = number_of_vertexes
        self.graph = []
        # construção da matriz de adjacência
        for i in range(number_of_vertexes):
            self.graph.append([0] * number_of_vertexes)
        # lista que armazena o número de vertices que estão interligados a partir de um vertiice especifico
        self.counter_list = [0] * number_of_vertexes

    # como o grafo é não direcionado a aresta interligam um ao o outro
    def connection(self,vertex1, vertex2):
        self.graph[vertex1 - 1][ vertex2 - 1] = 1
        self.graph[vertex2 - 1][vertex1 - 1] = 1

    # função de busca em profundidade
    def dfs(self, vertex):
        self.counter = 1
        # lista de vertices visitados
        self.visited = [0] * self.number_of_vertexes
        self.visited[vertex - 1] = 1
        # visita os vertices adjacentes
        self.visit(vertex)
        # adiciona à counter list o numero de vertices interligados entre si
        self.counter_list[vertex - 1] = self.counter

    # função que realiza a visita dos vertices adjacentes, incrementando na lista de visitados e o counter
    def visit(self, vertex):
        for i in range(self.number_of_vertexes):
            # visita caso vertice ainda não foi visitado
            if self.graph[vertex - 1][i] == 1 and self.visited[i] == 0:
                self.visited[i] = 1
                self.counter += 1
                self.visit(i + 1)


initial_input = input()
vertexes, connections_times = map(int, initial_input.split())
g = Graph(vertexes)

# adicionando as ligações (arestas)
for i in range(connections_times):
    connect = input()
    vertex1 , vertex2 = map(int, connect.split())
    g.connection(vertex1, vertex2)
# realizando a busca em profundidade
for j in range(vertexes):
    g.dfs(j + 1)
# output
print(*g.counter_list, sep=" ")
```
