```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Hash:
    def __init__(self, list, magic_number):
        self.hash_list = []
        self.hash_list_size = len(list)
        # cria os espaços no hash table
        for i in range(self.hash_list_size):
            self.hash_list.append(None)
        self.magic_number = magic_number

    def hash_function(self, char):
        return char % self.hash_list_size

    def insert(self, char):
        new_node = Node(char)
        hash_number = self.hash_function(char)
        # insere no espaço disponivel
        if not self.hash_list[hash_number]:
            self.hash_list[hash_number] = new_node
        # insere na lista encadeada se o espaço está indisponivel
        elif self.hash_list[hash_number]:
            aux_node = self.hash_list[hash_number]
            while aux_node.next is not None:
                aux_node = aux_node.next
            aux_node.next = new_node


    def cpf_check(self, list):
        for char in list:
            # procura a diferença entre o número e o magic number, caso esse número exista então a permissão será concedida, pois os dois números somados do hash table resultaram no magic number
            aux = self.magic_number - char
            if self.hash_list[self.hash_function(aux)] and self.hash_list[self.hash_function(aux)].value != char:
                if self.hash_list[self.hash_function(aux)].value == aux:
                    return print('UP Permission')
                elif self.hash_list[self.hash_function(aux)].next:
                    aux_node = self.hash_list[self.hash_function(aux)]
                    while aux_node.next is not None:
                        if aux_node.value == aux:
                            return print('UP Permission')
                        aux_node = aux_node.next

            # caso especial, caso tenha dois números iguais no hash table
            elif self.hash_list[self.hash_function(aux)] and self.hash_list[self.hash_function(aux)].value == char:
                if self.hash_list[self.hash_function(aux)].next:
                    aux_node = self.hash_list[self.hash_function(aux)]
                    while aux_node.next is not None:
                        if aux_node.value == aux:
                            return print('UP Permission')
                        aux_node = aux_node.next
        return print('NOT Permission')

def change_for_int(list):
    aux_list = []
    for char in list:
        aux_list.append(int(char) * 10)
    return aux_list


def check_duplicate(list):
    aux_list = []
    for char in list:
        if list.count(char) > 1:
            if char * list.count(char) not in aux_list:
                aux_list.append(char * list.count(char))
        else:
            aux_list.append(char)
    return aux_list

number_of_operations = int(input())
for i in range(number_of_operations):
    messages_input = input()
    cpf, magic_number = list(messages_input.split())
    magic_number = int(magic_number)
    # atualiza as strings da lista para inteiros e checa e soma os duplicados
    cpf = change_for_int(cpf)
    cpf_list = check_duplicate(cpf)

    hash_cpf = Hash(cpf, magic_number)
    # insere os números da lista na hash table
    for char in cpf_list:
        hash_cpf.insert(char)
    # checa permissão do cpf
    hash_cpf.cpf_check(cpf_list)
```

```python
class Hash_table:
    def __init__(self, hash_size):
        self.hash_list = []
        self.hash_list_size = hash_size
        # cria o tamanho do hash table
        for i in range(self.hash_list_size):
            self.hash_list.append(None)

    def hash_function(self, char):
        return char % self.hash_list_size

    def insert(self, char):
        hash_number = self.hash_function(char)
        # insere normalmente caso o espaço estiver disponivel
        if not self.hash_list[hash_number]:
            self.hash_list[hash_number] = char
            print(f'E: {hash_number}')
        else:
            # caso o lugar já esteja ocupado, irá procurar um lugar no hash table que esteja disponivel
            while self.hash_list[hash_number]:
                hash_number += 1
                if hash_number >= self.hash_list_size:
                    hash_number = 0
            self.hash_list[hash_number] = char
            print(f'E: {hash_number}')

    def search(self, char):
        hash_number = self.hash_function(char)
        if self.hash_list[hash_number]:
            # caso o número se encontra no espaço correto
            if self.hash_list[hash_number] == char:
                return print(f'E: {hash_number}')
            else:
                # caso não esteja no lugar do index, vai procurar onde possivelmente o número esteja na lista
                hash_number = 0
                while hash_number < self.hash_list_size:
                    if self.hash_list[hash_number] == char:
                        return print(f'E: {hash_number}')
                    hash_number += 1
            return print('NE')
        return print('NE')

    def query(self, char):
        # consulta a posição
        if self.hash_list[char]:
            return print(f'A: {self.hash_list[char]}')
        else:
            return print('D')

    def check_list(self):
        # checa possiveís espaços vazios, caso não exista o programa irá parar
        for char in self.hash_list:
            if char is None:
                return False
        print('Toda memoria utilizada')
        return True

hash_size = int(input())
number_of_operations = int(input())

hash_table = Hash_table(hash_size)
for i in range(number_of_operations):
    command, number = input().split()
    if command == 'ADD':
        hash_table.insert(int(number))
    elif command == 'SCH':
        hash_table.search(int(number))
    elif command == 'CAP':
        hash_table.query(int(number))
    if hash_table.check_list():
        break
```
