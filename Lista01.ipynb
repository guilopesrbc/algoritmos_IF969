```python
# cria o nó com dois ponteiros, escolhi usar a lista duplamente encadeada para facilitar na inversão da fila
class Node:
    def __init__(self, name, value):
        self.name = name
        self.value = float(value)
        self.next = None
        self.last = None

class Fila:
    def __init__(self, caixa):
        self.initial_node = None
        self.last_node = None
        self.caixa = int(caixa)
        self.dinheiro = 0
        self.size = 0

    def entrou(self, node):
        # seja o primeiro a entrar
        if self.initial_node is None:
            self.initial_node = node
            self.last_node = node
        # entra no final da fila
        else:
            self.last_node.next = node
            node.last = self.last_node
            self.last_node = node
        self.size += 1

    def saiu(self):
        actual_node = self.initial_node
        # caso só tenha uma pessoa da fila
        if actual_node.next is None:
            self.dinheiro += actual_node.value
            self.initial_node = None
            self.last_node = None
        # sai o primeiro da fila e arrecada o dinheiro pago
        else:
            self.dinheiro += actual_node.value
            self.initial_node = actual_node.next
            actual_node.next.last = None
        self.size -= 1
        print(f'{actual_node.name} foi chamado para o caixa {self.caixa}')

    # função que exclui o ultimo da fila para auxiliar na função inverter
    def excluir_no_final(self):
        actual_node = self.last_node
        # caso só reste uma pessoa na fila
        if actual_node == self.initial_node:
            self.last_node = None
            self.initial_node = None
            self.size -= 1
        else:
            self.last_node = actual_node.last
            self.last_node.next = None
            self.size -= 1

    def tamanho_fila(self):
        return self.size

    def meio_fila(self):
        return round((self.tamanho_fila() / 2)+0.5)

    def inverteu(self):
        meio = self.meio_fila()
        # cria uma nova fila com o número do caixa que está vazio
        if self.caixa == 1:
            nova_fila = Fila(2)
        else:
            nova_fila = Fila(1)
        actual_node = self.last_node
        # adiciona as pessoas na nova fila de forma invertida e exclui os que foram adicionados na outra fila
        for count in range(meio):
            node = Node(actual_node.name, actual_node.value)
            nova_fila.entrou(node)
            self.excluir_no_final()
            actual_node = actual_node.last
        return nova_fila

    def arrecadado(self):
        return self.dinheiro

fila_um = Fila(1)
fila_dois = Fila(2)

while True:
    entrada = input().split()

    # comando entrou na fila
    if entrada[0] == "ENTROU:":
        # caixa 1
        if entrada[2] == "1":
            new_node = Node(entrada[1], entrada[3])
            fila_um.entrou(new_node)
            print(f'{new_node.name} entrou na fila {fila_um.caixa}')
        # caixa 2
        else:
            new_node = Node(entrada[1], entrada[3])
            fila_dois.entrou(new_node)
            print(f'{new_node.name} entrou na fila {fila_dois.caixa}')

    # comando proximo na fila
    elif entrada[0] == "PROXIMO:":
        # caixa 1
        if entrada[1] == "1":
            if fila_um.tamanho_fila() == 0:
                pago = fila_um.arrecadado()
                nova_fila = fila_dois.inverteu()
                # adiciona a nova fila ao caixa que foi chamado
                fila_um = nova_fila
                fila_um.dinheiro += pago
                fila_um.saiu()
            else:
                fila_um.saiu()
        # caixa 2
        else:
            if fila_dois.tamanho_fila() == 0:
                pago = fila_dois.arrecadado()
                nova_fila = fila_um.inverteu()
                # adiciona a nova fila ao caixa que foi chamado
                fila_dois = nova_fila
                fila_dois.dinheiro += pago
                fila_dois.saiu()
            else:
                fila_dois.saiu()
    # comando FIM
    elif entrada[0] == "FIM":
        print(f'Caixa 1: R$ {fila_um.arrecadado():.2f}, Caixa 2: R$ {fila_dois.arrecadado():.2f}')
        break
```

```python
class Node:
    def __init__(self, node, posicao):
        self.value = node
        self.posicao = posicao
        self.next = None

class Pilha:
    def __init__(self):
        self.initial_node = None
        self.last_node = None
        # para checar se finzalizou com mais V ou mais F
        self.fim = False
        self.size = 0

    def inserir(self, node):
        # seja o primeiro a entrar
        if self.initial_node is None:
            self.initial_node = node
            self.last_node = node
        # entra no final da pilha
        else:
            self.last_node.next = node
            self.last_node = node
        self.size += 1

    def excluir(self):
        actual_node = self.initial_node
        self.initial_node = actual_node.next
        self.size -= 1

    def tamanho(self):
        return self.size


entrada = input()
pilha = Pilha()
for posicao, char in enumerate(entrada):
    # adiciona o F na pilha, juntamente com sua posição
    if char == "F":
        new_node = Node(char, posicao)
        pilha.inserir(new_node)
    elif char == "V":
        # confere se tinha F para ser fechado
        if pilha.tamanho() == 0:
            print(f'Incorreto, devido a capa na posição {posicao + 1}.')
            pilha.fim = True
            break
        pilha.excluir()
# caso tenha finalizado com mais F do que V
if pilha.tamanho() == 0 and pilha.fim == False:
    print('Correto.')
elif pilha.fim == False:
    print(f'Incorreto, devido a capa na posição {pilha.initial_node.posicao + 1}.')
```

```python
# cria o nó
class Node:
    def __init__(self, codigo):
        self.value = int(codigo)
        self.next = None


class Pilha:
    def __init__(self):
        self.initial_node = None
        self.last_node = None

    # função para inserir a pilha inicial conforme vem no input
    def pilha_inicial(self, node):
        # seja o primeiro a entrar
        if self.initial_node is None:
            self.initial_node = node
            self.last_node = node
        # entra no final da pilha
        else:
            self.last_node.next = node
            self.last_node = node

    # insere o novo codigo na pilha
    def inserir(self, node):
        actual_node = self.initial_node
        while True:
            if node.value >= actual_node.value and node.value < actual_node.next.value:
                node.next = actual_node.next
                actual_node.next = node
                break
            actual_node = actual_node.next

    # função printar retorna a pilha em forma de lista para o output
    def printar(self):
        actual_node = self.initial_node
        lista_aux = []
        lista_aux.append(str(actual_node.value))
        while actual_node != self.last_node:
            actual_node = actual_node.next
            lista_aux.append(str(actual_node.value))
        return lista_aux


def pilhaImaculada(pilha):
    actual_node = pilha.initial_node
    while actual_node.value != pilha.last_node.value:
        if actual_node.next.value >= actual_node.value:
            actual_node = actual_node.next
        else:
            print("A pilha está um caos.")
            return False
    return True


def novaLocacao(pilha, codigo):
    if pilhaImaculada(pilha) is True:
        new_node = Node(codigo)
        pilha.inserir(new_node)
        print(pilha.printar())


entrada = input().split(",")
codigo = int(input())

pilha = Pilha()
# adiciona o input inicial na pilha
for char in entrada:
    new_node = Node(char)
    pilha.pilha_inicial(new_node)

novaLocacao(pilha, codigo)
```

```python
# cria o nó com dois ponteiros
class Node:
    def __init__(self, node):
        self.value = node
        self.next = None
        self.last = None

class Lista_dupla:
    def __init__(self, node):
        self.initial_node = node
        self.last_node = None
        node.next = None
        node.last = None

    def add(self, node):
        # caso não tenha nenhum nó
        if self.initial_node is None:
            self.initial_node = node
            self.last_node = node
            self.initial_node.next = None
            node.next = None
            node.last = None
        # se tiver um nó na lista
        elif self.initial_node.next is None:
            aux = self.initial_node
        # ponteiros primeiro nó
            self.initial_node.next = node
            aux.next = node
            self.last_node = node
        # definição de ponteiros
            node.last = self.initial_node
            node.next = None
        else:
            aux = self.last_node
        # definição ponteiros
            aux.next = node
            node.next = None
            node.last = self.last_node
            self.last_node = node

    def rem(self, node):
        # caso seja o primeiro e unico
        if node == self.initial_node and self.initial_node.next is None:
            self.initial_node = None
        # caso seja o primeiro
        elif node == self.initial_node and self.initial_node.next is not None:
            node.next.last = None
            self.initial_node = node.next
        # caso seja o ultimo
        elif node == self.last_node:
            node.last.next = None
            node.last = self.last_node
        else:
            node.last.next = node.next
            node.next.last = node.last

    def exib(self):
        actual_node = self.last_node
        if actual_node is not None:
            print(actual_node.value)
            while actual_node.last is not None:
                actual_node = actual_node.last
                print(actual_node.value)
        # caso não tenha nada na lista
        else:
            return None

    def search_obj(self,chave):
        actual_node = self.initial_node
        while actual_node is not None and actual_node.value != chave:
            actual_node = actual_node.next
        return actual_node

    def find(self, node):
        # removendo
        self.rem(node)
       # adicionando
        self.add(node)


# primeira entrada para chamar a classe
primeiro_add = input().split()
node1 = Node(primeiro_add[1])
if primeiro_add[0] == "ADD":
    lista = Lista_dupla(node1)

end = False
while end is False:
    entrada = input().split(" ")
    if entrada[0] == "END":
        end = True
    elif entrada[0] == "ADD":
        node = Node(entrada[1])
        lista.add(node)
    elif entrada[0] == "REM":
        # para remoção preciso do objeto para ajustar os ponteiros do nó proximo e antes dele, caso ele n esteja no final nem no inicio da lista
        node = lista.search_obj(entrada[1])
        lista.rem(node)
    elif entrada[0] == "FIND":
        # mesma lógica da remoção, precisa do objeto para remover
        node = lista.search_obj(entrada[1])
        lista.find(node)
    elif entrada[0] == "EXIB":
        lista.exib()
```
