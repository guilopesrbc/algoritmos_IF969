```python
class Sort():
    def __init__(self, name):
        self.comparations = 0
        self.swaps = 0
        self.name = name

    def bubblesort(self,array, actions):
        self.comparations = 0
        self.swaps = 0
        for j in range(len(array)):
            # compara o valor com seu sucessor e realiza as trocas caso o valor seja maior
            for i in range(0, len(array) - j - 1):
                self.comparations += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array
                if array[i] > array[i + 1]:
                    temp = array[i]
                    array[i] = array[i + 1]
                    array[i + 1] = temp
                    self.swaps += 1
                    if actions > 0 and (self.comparations + self.swaps) == actions:
                        return array

    def selectionsort(self, array, actions):
        self.comparations = 0
        self.swaps = 0
        for step in range(len(array)):
            low_idx = step
            # loop para achar o valor minimo
            for i in range(step + 1, len(array)):
                if array[i] < array[low_idx]:
                    low_idx = i
                self.comparations += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array

            # efetua a troca do valor minimo para sua posição (step) adequada
            if low_idx != step:
                temp = array[step]
                array[step] = array[low_idx]
                array[low_idx] = temp
                self.swaps += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array

    def insertionsort(self, array, actions):
        self.comparations = 0
        self.swaps = 0
        # começa do segundo elemento da lista e vai comparando com os anteriores, se repetindo até o final da lista
        for step in range(1, len(array)):
            key = array[step]
            before = step - 1

            # compara com os anteriores até haver um valor menor ou chegar ao inicio do array
            while before >= 0 and key < array[before]:
                self.comparations += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array
                array[before + 1] = array[before]
                before -= 1
                self.swaps += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array

            # pôe o valor no lugar correto
            if before != step - 1:
                array[before + 1] = key
            # caso chegue no primeiro valor do array, vai fazer mais uma comparação
            if before != -1:
                self.comparations += 1
                if actions > 0 and (self.comparations + self.swaps) == actions:
                    return array


    def shellsort(self, array, actions):
        self.comparations = 0
        self.swaps = 0
        # calcula o gap inicial
        interval = len(array) // 2
        while interval > 0:
            # começa as divisões de acordo com o gap
            for i in range(interval, len(array)):
                temp = array[i]  # valor a ser comparado
                j = i  # index anterior de acordo com o gap

                # compara com os index do gap anterior até achar o lugar adequado do valor temp
                while j >= interval and array[j - interval] > temp:
                    self.comparations += 1
                    if actions > 0 and (self.comparations + self.swaps) == actions:
                        return array
                    array[j] = array[j - interval]
                    j -= interval
                    self.swaps += 1
                    if actions > 0 and (self.comparations + self.swaps) == actions:
                        return array

                # troca o valor para o lugar adequado
                array[j] = temp

                # caso o numero anterior não seja menor, conta uma comparação que não foi feita no loop
                if j >= interval:
                    self.comparations += 1
                    if actions > 0 and (self.comparations + self.swaps) == actions:
                        return array
            interval //= 2

    def quicksort(self, array, low_idx, high_idx):
        # divide as partições e chama a função para as trocas
        if low_idx < high_idx:
            pivot = self.partition(array, low_idx, high_idx)
            self.quicksort(array, low_idx, pivot)
            self.quicksort(array, pivot + 1, high_idx)

    def partition(self, array, low_idx, high_idx):
        # determina pivot e index i e j
        pivot = array[(high_idx + low_idx) // 2]
        i = low_idx
        j = high_idx
        while True:
            # condição de parada, quando i passou o j, pivot achou a posição correta
            if i >= j:
                return j

            # procura os valores maiores que o pivot
            while array[i] < pivot:
                self.comparations += 1
                i += 1
            # procura os valores menores que o pivot
            while array[j] > pivot:
                self.comparations += 1
                j -= 1
            # realiza troca entre o valor do index i e o index j
            temp = array[i]
            array[i] = array[j]
            array[j] = temp
            self.swaps += 1


list_of_input = list(map(int, input().split()))
bubble = Sort('Caça-Rato')
selection = Sort('Grafite')
insertion = Sort('Lacraia')
shell = Sort('Rivaldo')
quick = Sort('Toninho')

# bubble sort
bubble.bubblesort(list_of_input[:], 0)
bubble_sum = bubble.comparations + bubble.swaps
print(f'{bubble.name} ordena a lista com {bubble.comparations} comparações e {bubble.swaps} trocas.')

# selection sort
selection.selectionsort(list_of_input[:], 0)
selection_sum = selection.comparations + selection.swaps
print(f'{selection.name} ordena a lista com {selection.comparations} comparações e {selection.swaps} trocas.')

# insertion sort
insertion.insertionsort(list_of_input[:], 0)
insertion_sum = insertion.comparations + insertion.swaps
print(f'{insertion.name} ordena a lista com {insertion.comparations} comparações e {insertion.swaps} trocas.')

# shell sort
shell.shellsort(list_of_input[:], 0)
shell_sum = shell.comparations + shell.swaps
print(f'{shell.name} ordena a lista com {shell.comparations} comparações e {shell.swaps} trocas.')

# quick sort
quick.quicksort(list_of_input[:], 0, len(list_of_input[:]) - 1)
quick_sum = quick.comparations + quick.swaps
print(f'{quick.name} ordena a lista com {quick.comparations} comparações e {quick.swaps} trocas.')

# resto do programa
print("-VENCEDOR DA RODADA-")

if min(bubble_sum, selection_sum, insertion_sum, shell_sum, quick_sum) == bubble_sum:
    print(f'O vencedor da rodada é {bubble.name}, com {bubble_sum} ações.')
    print('-Toninho está a dormir...-')
    print('Os outros estagiários retornam as seguintes listas com essa quantidade de ações:')
    print(f'Com {bubble_sum} ações, {selection.name} ordena a lista assim: {selection.selectionsort(list_of_input[:], bubble_sum)}')
    print(f'Com {bubble_sum} ações, {insertion.name} ordena a lista assim: {insertion.insertionsort(list_of_input[:], bubble_sum)}')
    print(f'Com {bubble_sum} ações, {shell.name} ordena a lista assim: {shell.shellsort(list_of_input[:], bubble_sum)}')

elif min(bubble_sum, selection_sum, insertion_sum, shell_sum, quick_sum) == selection_sum:
    print(f'O vencedor da rodada é {selection.name}, com {selection_sum} ações.')
    print('-Toninho está a dormir...-')
    print('Os outros estagiários retornam as seguintes listas com essa quantidade de ações:')
    print(f'Com {selection_sum} ações, {bubble.name} ordena a lista assim: {bubble.bubblesort(list_of_input[:], selection_sum)}')
    print(f'Com {selection_sum} ações, {insertion.name} ordena a lista assim: {insertion.insertionsort(list_of_input[:], selection_sum)}')
    print(f'Com {selection_sum} ações, {shell.name} ordena a lista assim: {shell.shellsort(list_of_input[:], selection_sum)}')

elif min(bubble_sum, selection_sum, insertion_sum, shell_sum, quick_sum) == insertion_sum:
    print(f'O vencedor da rodada é {insertion.name}, com {insertion_sum} ações.')
    print('-Toninho está a dormir...-')
    print('Os outros estagiários retornam as seguintes listas com essa quantidade de ações:')
    print(f'Com {insertion_sum} ações, {bubble.name} ordena a lista assim: {bubble.bubblesort(list_of_input[:], insertion_sum)}')
    print(f'Com {insertion_sum} ações, {selection.name} ordena a lista assim: {selection.selectionsort(list_of_input[:], insertion_sum)}')
    print(f'Com {insertion_sum} ações, {shell.name} ordena a lista assim: {shell.shellsort(list_of_input[:], insertion_sum)}')

elif min(bubble_sum, selection_sum, insertion_sum, shell_sum, quick_sum) == shell_sum:
    print(f'O vencedor da rodada é {shell.name}, com {shell_sum} ações.')
    print('-Toninho está a dormir...-')
    print('Os outros estagiários retornam as seguintes listas com essa quantidade de ações:')
    print(f'Com {shell_sum} ações, {bubble.name} ordena a lista assim: {bubble.bubblesort(list_of_input[:], shell_sum)}')
    print(f'Com {shell_sum} ações, {selection.name} ordena a lista assim: {selection.selectionsort(list_of_input[:], shell_sum)}')
    print(f'Com {shell_sum} ações, {insertion.name} ordena a lista assim: {insertion.insertionsort(list_of_input[:], shell_sum)}')

elif min(bubble_sum, selection_sum, insertion_sum, shell_sum, quick_sum) == quick_sum:
    print(f'O vencedor da rodada é {quick.name}, com {quick_sum} ações.')
    print('-Toninho está a dormir...-')
    print('Os outros estagiários retornam as seguintes listas com essa quantidade de ações:')
    print(f'Com {quick_sum} ações, {bubble.name} ordena a lista assim: {bubble.bubblesort(list_of_input[:], quick_sum)}')
    print(f'Com {quick_sum} ações, {selection.name} ordena a lista assim: {selection.selectionsort(list_of_input[:], quick_sum)}')
    print(f'Com {quick_sum} ações, {insertion.name} ordena a lista assim: {insertion.insertionsort(list_of_input[:], quick_sum)}')
    print(f'Com {quick_sum} ações, {shell.name} ordena a lista assim: {shell.shellsort(list_of_input[:], quick_sum)}')
```

```python
def mergesort(array):
    if len(array) > 1:
        mid = len(array) // 2
        # divide o array em duas partições
        sub_array1 = array[:mid]
        sub_array2 = array[mid:]
        # realiza a divisão e o merge das partições de forma recursiva até dividir todas as partições
        mergesort(sub_array1)
        mergesort(sub_array2)
        if sub_array1 == sub_array2:
            pass
        else:
            merge(array, sub_array1, sub_array2) # realiza o merge das partições de forma recursiva
            # dessa forma, o primeiro merge será feito quando as partições tiverem 2 valores somente
            # e em seguida fazendo o merge das demais partições

def merge(array, sub_array1, sub_array2):
    idx_sub1 = 0 # index sub array 1
    idx_sub2 = 0 # index sub array 2
    idx = 0 # index array

    # realiza o merge entre as duas partições, compara os valores de cada partição, realiza a modificação no merge list (array) e incrementa o index que foi modificado
    while idx_sub1 < len(sub_array1) and idx_sub2 < len(sub_array2):
        if sub_array1[idx_sub1] < sub_array2[idx_sub2]:
            array[idx] = sub_array1[idx_sub1]
            idx_sub1 += 1

        else:
            array[idx] = sub_array2[idx_sub2]
            idx_sub2 += 1
        idx += 1

    # caso o sub array 1 não fez totalmente o merge no array
    while idx_sub1 < len(sub_array1):
        array[idx] = sub_array1[idx_sub1]
        idx_sub1 += 1
        idx += 1

    # caso o sub array 2 não fez totalmente o merge no array
    while idx_sub2 < len(sub_array2):
        array[idx] = sub_array2[idx_sub2]
        idx_sub2 += 1
        idx += 1

def change_for_int(array):
    for i in range(len(array)):
        array[i] = int(array[i])
    return array

array = change_for_int(input().split() + input().split())
mergesort(array)

# calcula a mediana e printa
if len(array) % 2 == 0:
    median = (array[len(array)//2 -1] + array[len(array)//2 ]) / 2
    print(f'O salário sugerido por Juba na primeira negociação será de {median:.2f} mil reais.')
else:
    median = array[len(array)//2]
    print(f'O salário sugerido por Juba na primeira negociação será de {median:.2f} mil reais.')
```

```python
def max_heapify(array,index):
    # verifica quais são os filhos do carácter atual
    left_child_index = left(index)
    right_child_index = right(index)

    # caso exista, o filho da esquerda seja maior que o carácter atual
    if left_child_index < len(array) and array[left_child_index] > array[index]:
        largest = left_child_index

    else:
        largest = index

    # caso exista, o filho da direita seja maior que o carácter atual ou maior que o filho à esquerda
    if right_child_index < len(array) and array[right_child_index] > array[largest]:
        largest = right_child_index

    # caso algum dos filhos sejam maior, realiza a troca e continua o heapify do maior até ele chegar no seu lugar adequado
    if largest != index:
        temp = array[index]
        array[index] = array[largest]
        array[largest] = temp
        max_heapify(array, largest)

def left(index):
    return 2 * index + 1

def right(index):
    return 2 * index + 2

def build_max_heap(array):
    # realiza o heap de maximo começando pela metade do array e realizando um down até à primeira posição. PS: começa da metade, pois nenhuma troca é realizada no heapify até o index do meio, então o programa realiza mais comandos desnecessários.
    for index in range(len(array)//2 -1, -1, -1):
        max_heapify(array,index)

def extract_max(array):
    # troca o maior valor pelo menor valor e realiza o heap de maximo novamente
    max = array[0]
    array[0] = array[-1]
    array.pop(-1)
    max_heapify(array, 0)
    return max

def min_heap(array):
    min = array[len(array)-1]
    for i in range(len(array) -1,len(array)//2 -1, -1):
        if array[i] < min:
            min = array[i]
    return min


def change_for_int(array):
    for i in range(len(array)):
        array[i] = int(array[i])
    return array

sequence = change_for_int(input().split())
constant = int(input())
counter = 0
while len(sequence)!= 0:
    # realiza o heap de maximo
    build_max_heap(sequence)
    # exclui o maior valor do heap
    maximum = extract_max(sequence)
    if len(sequence) != 0:
        minimum = min_heap(sequence)
        k = maximum - abs(minimum * constant)
        if k > 0:
            sequence.append(k)
    counter += 1
print(f'{counter} rodadas, partindo para a próxima!')
```
